```python
import requests
from bs4 import BeautifulSoup
import time
```

```python
urls =[]  # "https://www.mayoclinic.org/symptoms/index?letter=A"
x = "A"
for i in range(26):
    urls.append(f"https://www.mayoclinic.org/symptoms/index?letter={chr(ord(x) + i)}")
```

```python
urls
```

```python
for url in urls:
    try:
        a = url[-1]
        r = requests.get(url)
        print(r)
        htmlContent = r.content
        soup = BeautifulSoup(htmlContent, 'html.parser')
        print(url[-1])
        symptoms= (soup.find(string=a).findNext("ol"))
        symptom_list = symptoms.text
        print(symptom_list)
        
        #open text file
        text_file = open('C:/Users/91966/Desktop/trias/symptoms.txt', "a")
        text_file.write(url[-1])
        text_file.write(symptom_list)
        

#         time.sleep(2)
    except AttributeError:
        pass
```
